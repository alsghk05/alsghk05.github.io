```python
# '접수일시', '배차일시', '승차일시', '하차일시' 열을 datetime 형식으로 변환
df['접수일시'] = pd.to_datetime(df['접수일시'])
df['배차일시'] = pd.to_datetime(df['배차일시'])
df['승차일시'] = pd.to_datetime(df['승차일시'])
df['하차일시'] = pd.to_datetime(df['하차일시'])

df['배차대기시간'] = (df['배차일시'] - df['접수일시']).dt.total_seconds() / 60
df['승차대기시간'] = (df['승차일시'] - df['배차일시']).dt.total_seconds() / 60
df['총대기시간'] = (df['승차일시'] - df['접수일시']).dt.total_seconds() / 60
df['승차시간'] = (df['하차일시'] - df['승차일시']).dt.total_seconds() / 60

# 시:분:초로 변환
df['배차대기시간'] = pd.to_datetime(df['배차대기시간'], unit='m').dt.strftime('%H:%M:%S')
df['승차대기시간'] = pd.to_datetime(df['승차대기시간'], unit='m').dt.strftime('%H:%M:%S')
df['총대기시간'] = pd.to_datetime(df['총대기시간'], unit='m').dt.strftime('%H:%M:%S')
df['승차시간'] = pd.to_datetime(df['승차시간'], unit='m').dt.strftime('%H:%M:%S')

df[['배차대기시간', '승차대기시간', '승차시간', '총대기시간']]
```

* 호출지와 목적지 사이의 직선거리 구하기

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # 지구의 반지름 (km 단위)
    R = 6371.0

    # 라디안으로 변환
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    # 위도 및 경도 간의 차이
    dlon = lon2 - lon1
    dlat = lat2 - lat1

    # Haversine 공식 적용
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    # 거리 계산
    distance = R * c
    return distance

# 직선 거리를 계산하여 새로운 열에 추가
df['직선거리'] = df.apply(lambda row: calculate_distance(row['호출지위도'], row['호출지경도'], row['목적지위도'], row['목적지경도']), axis=1)

print(df[['호출지경도', '호출지위도', '목적지경도', '목적지위도', '직선거리']])
```

```python
# 직선거리를 구간별로 나누기
distance_bins = [0, 3, 6, 9, 12, 15, 18, 21, np.inf]
distance_labels = ['0-3', '3-6', '6-9', '9-12', '12-15', '15-18', '18-21', '21+']
df['직선거리구간'] = pd.cut(df['직선거리(km)'], bins=distance_bins, labels=distance_labels, right=False)

# 각 구간별로 배차대기시간의 평균 계산
waiting_time_by_distance = df.groupby('직선거리구간')['배차대기시간(분)'].mean()

# 시각화
plt.figure(figsize=(10, 6))
waiting_time_by_distance.plot(kind='bar', color='skyblue')
plt.title('직선거리별 평균 배차대기시간')
plt.xlabel('직선거리(km)')
plt.ylabel('평균 배차대기시간(분)')
plt.xticks(rotation=45)
plt.show()
```

```python
# 직선거리와 배차대기시간 간의 상관 관계 분석
correlation_distance_waiting_time = df[['직선거리(km)', '배차대기시간(분)']].corr().iloc[0, 1]

print("직선거리와 배차대기시간 간의 상관 계수:", correlation_distance_waiting_time)
```

```python
# 구간에 속하는 데이터 할당
df['distance_bin'] = pd.cut(df['직선거리(km)'], bins=distance_bins)

# 각 구간의 빈도 계산
distance_frequency = df['distance_bin'].value_counts().sort_index()

print(distance_frequency)
```

```python
# '직선거리(km)'의 최대값을 포함하도록 bins와 labels 수정
max_distance = df['직선거리(km)'].max()
bins = [0, 3000, 6000, 9000, 12000, 15000, 18000, 21000, 24000, max_distance]
labels = ['0-3km', '3-6km', '6-9km', '9-12km', '12-15km', '15-18km', '18-21km', '21-24km', f'{max_distance:.0f}km+']

# 거리그룹 생성
df['거리그룹'] = pd.cut(df['직선거리(km)'], bins=bins, labels=labels)

# 거리그룹별 총 대기시간 데이터 준비
distance_groups = df.groupby('거리그룹')['총대기시간(분)'].apply(list)

# 거리그룹별 총 대기시간 분포 시각화
plt.figure(figsize=(12, 6))
df.boxplot(column='총대기시간(분)', by='거리그룹', grid=False)
plt.title('총 대기시간 분포 by 거리그룹')
plt.suptitle('')
plt.xlabel('거리그룹')
plt.ylabel('총 대기시간(분)')
plt.show()

# ANOVA 검정 수행
anova_result_distance = stats.f_oneway(*distance_groups)

print(f"거리그룹 별 ANOVA 결과: F-statistic = {anova_result_distance.statistic}, p-value = {anova_result_distance.pvalue}")
```
