```python
import pandas as pd

# 파일 로드
train_data = pd.read_csv('train.csv')

# 데이터의 처음 몇 줄을 확인
train_data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 수치형 데이터만을 포함하는 컬럼 선택
numerical_columns = train_data.select_dtypes(include=['float64', 'int64']).columns

# 상관계수 계산
correlation_matrix = train_data[numerical_columns].corr()

# 상관계수 히트맵 시각화
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Numerical Features Correlation Matrix")
plt.show()
```

```python
# train.csv 데이터의 컬럼 목록 확인
train_data.columns
```

```python
# 분석에 포함할 변수들 선택
selected_columns = ['BREADTH', 'BUILT', 'DEADWEIGHT', 'GT', 'LENGTH', 'PORT_SIZE', 'ATA_LT']

# 선택된 변수들에 대한 상관관계 계산
selected_correlation_matrix = train_data[selected_columns].corr()

# 상관계수 히트맵 시각화
plt.figure(figsize=(10, 6))
sns.heatmap(selected_correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Selected Features Correlation Matrix")
plt.show()
```
