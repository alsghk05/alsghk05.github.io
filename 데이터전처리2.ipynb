* 호출지와 목적지 사이의 직선거리 구하기

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # 지구의 반지름 (km 단위)
    R = 6371.0

    # 라디안으로 변환
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    # 위도 및 경도 간의 차이
    dlon = lon2 - lon1
    dlat = lat2 - lat1

    # Haversine 공식 적용
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    # 거리 계산
    distance = R * c
    return distance

# 직선 거리를 계산하여 새로운 열에 추가
df['직선거리(km)'] = df.apply(lambda row: calculate_distance(row['호출지위도'], row['호출지경도'], row['목적지위도'], row['목적지경도']), axis=1)

df[['호출지경도', '호출지위도', '목적지경도', '목적지위도', '직선거리(km)']]
```
